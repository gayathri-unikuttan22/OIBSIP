```python
import numpy as npy
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interact
%matplotlib inline
```

```python
col=['sepal length','sepal width','petal length','petal width','species']
dfn=pd.read_csv('Iris.csv',names=col,skiprows=1)
dfn.head(10)
```

```python
dfn.describe()
```

```python
sns.pairplot(dfn ,hue='species')
```

```python
datam=dfn.values;
X=datam[:,0:4]
Y=datam[:,4]
#print(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.5)
print(X_train)
```

```python
from sklearn.linear_model import LogisticRegression
model_lr=LogisticRegression()
model_lr.fit(X_train,Y_train)
```

```python
pred1=model_lr.predict(X_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,pred1)*100)
for i in range(len(pred1)):
  print(Y_test[i],pred1[i])
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,pred1))
```
